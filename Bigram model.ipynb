```python
import nltk
import random
from nltk.util import bigrams
from nltk.corpus import reuters
import json
```

```python
def preprocess_abstract(abstract):
    # Tokenize the abstract
    tokens = nltk.word_tokenize(abstract.lower())
    # Create bigrams
    return list(bigrams(tokens))
# Reading file 
j_file = open("./training_data_final.json",'r',encoding="utf-8")
data = json.loads(j_file.read())

# Reading abstracts
abstracts = [data[key]["abstract"] for key in data]
bigram_abstracts = [preprocess_abstract(abstract) for abstract in abstracts]
```

```python
def build_bigram_model(abstracts):
    # Combine all bigrams from abstracts into a single list
    all_bigrams = [bigram for abstract in abstracts for bigram in abstract]
    # Create a frequency distribution of bigrams
    bigram_freq_dist = nltk.FreqDist(all_bigrams)
    # Create a conditional frequency distribution of bigrams
    bigram_cfd = nltk.ConditionalFreqDist(all_bigrams)
    return bigram_freq_dist, bigram_cfd

bigram_freq_dist, bigram_cfd = build_bigram_model(bigram_abstracts)
```

```python
def generate_abstract(title, num_words=100):
    # Tokenize the title
    title_tokens = nltk.word_tokenize(title.lower())
    # Initialize the abstract with the title tokens
    abstract_tokens = title_tokens.copy()
    
    # Generate the rest of the abstract using the bigram model
    for _ in range(num_words):
        # Get the last word in the abstract_tokens
        last_word = abstract_tokens[-1]
        # Get the most likely next word based on the bigram model
        next_word = random.choice(list(bigram_cfd[last_word].keys()))
        abstract_tokens.append(next_word)
    
    # Combine the tokens into a single abstract string
    generated_abstract = " ".join(abstract_tokens)
    return generated_abstract

# Example usage:
titles = [data[key]["title"] for key in data]
title_list = []
for title in titles:
    title_list.append(title)
    
generated_abstract = generate_abstract(title)
print(f"Title: {title_list[0]}")
print(f"Generated Abstract: {generated_abstract}\n")

generated_abstract = generate_abstract('Specific attention is all you need')
print(f"Title: {'Specific attention is all you need'}")
print(f"Generated Abstract: {generated_abstract}\n")

```
